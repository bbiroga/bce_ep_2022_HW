### Figure 3

```python
#Creating Figure 3
dv.creating_figure_three(nodes, G)
```

**Figure 3** shows the correlation between demographics and network characteristics (Embededness was out of scope for this project)
* **(a) Degree Centrality:** We can see the number of average connections for age groups by gender. As we can see younger people are engaged more in this social media platform, the number of connections peak around age 20. From that point its rapidly decreasing for both genders. Comparing to the results of the original article we see a similar pattern, but without the rebound at middle aged people (but keep in mind that our data is also shorter).

* **(b) Neighbor Connectivity:** Shows the average degree of neighbors of a specific user. The dynamic is the same as in degree of centrality but male users have a higher connectivity until the age of 40. We conclude that from the age of 40 female users tend to reconnect to the network (probalby through their children). This pattern is similar to the one in Dong et al 2014.

* **(c) Triadic Closure:** Triadic closure shows the local clustering coefficient (which shows how closed friend groups are). Interestengly it is different from the original article. Young people have high clustering effect but its decreasing until their mid 30s (probably because on social media in a friend group everybody connects to each other). The clustering increases after the age of 35 especially for females, this could be explained by parents getting to know their children's parents too.


### Figure 5

```python
#Adding needed feature to edges and creating Figure 5
edges_w_features = gm.add_node_features_to_edges(nodes, edges)
dv.creating_figure_five(edges_w_features)
```

**Figure 5** shows 4 heatmaps of the strenghts of the social ties for different pairs. We conclude the followings:

* We can see general homophily: specific users strongest ties tend to be with people of the same age

* Ligther yellow to green areas show the cross-generational connections which is the widest for Female-Female pairs, probably through motherh-daughter and daughter's friends connections

* On Male-Male connections the bottom left red area is stronger compared to Female-Female, therefore younger males tend to have wider age range for their friend group while females only have connections closer to their own age (again homophily)

* On the Male-Female pairs we can see patterns of the dating and reproductively active period: it is assymmetrical, Men tend to have more connections with younger Women
